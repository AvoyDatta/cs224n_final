```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
attn_vectors = np.load('attn_weights.npy')
print(attn_vectors.shape)
t = np.linspace(1,10,500)
labels = [" 4 days ago ", "3 days ago", "2 days ago", "1 day ago", "today"]

plt.plot(t,attn_vectors[:,:,0],t,attn_vectors[:,:,1],t,attn_vectors[:,:,2],t,attn_vectors[:,:,3],t,attn_vectors[:,:,4])
plt.legend(labels)
plt.save('attn_plt.png')
```
